### 计算vocab-dist

```python
# 计算vocab-dist
from math import radians, cos, sin, asin, sqrt
from scipy.spatial.distance import  cdist

def geo_distance(c_i, c_j):
    lng1, lat1 = c_i[0], c_i[1]
    lng2, lat2 = c_j[0], c_j[1]
    #lng1,lat1,lng2,lat2 = (120.12802999999997,30.28708,115.86572000000001,28.7427)
    lng1, lat1, lng2, lat2 = map(radians, [float(lng1), float(lat1), float(lng2), float(lat2)]) # 经纬度转换成弧度
    dlon=lng2-lng1
    dlat=lat2-lat1
    a=sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    distance=2*asin(sqrt(a))*6371*1000 # 地球平均半径，6371km
    distance=round(distance,3)
    # print(c_i, c_j, distance)
    return distance
```

```python
%%time
dataset_name = 'CDR'
loc_index_cdr = pd.read_csv('./CDR/processed/loc_index_cdr_582.csv')
loc_a = loc_index_cdr[['longitude','latitude']].values.tolist()
loc_b = loc_a
dm = cdist(loc_a, loc_b, lambda loc_a, loc_b: geo_distance(loc_a, loc_b))
dm = pd.DataFrame(dm, columns=[i for i in range(len(loc_a))])
dm.index = loc_index_cdr.index
dm.index.name = 'loc_index'
dm.to_hdf(f'./{dataset_name}/processed/vocab-dist.h5', key='data')
```

```python
%%time
V = [np.array(10*[i]) for i in range(4)]
D = 4*[np.zeros(10)]
for i in range(len(dm)):
    print(f'Traj: {i}')
    tmp = dm[i].sort_values()[:10]
    v = tmp.index.to_numpy()
    d = tmp.values
    V.append(v)
    D.append(d)
V, D = np.array(V), np.array(D)
```
